```python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data
casino = 'https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/JitteredHeadCount.csv'
casino_data = pd.read_csv(casino)
casino_data.describe()
```

```python
#observations from summary statistics
# 1. Maximum number of poeple at any given time has been 109. so the casino can plan their staffing requirements accordingly. 
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
ax = plt.figure(figsize=(6, 6)).gca() # define axis
casino_data.plot.scatter(x = 'HeadCount', y = 'Hour', ax = ax)
ax.set_title('Hour vs Headcount') # Give the plot a main title
ax.set_ylabel('Hour')# Set text for y axis
ax.set_xlabel('HeadCount')
```

```python
# there is a major dip in the casino traffic from 5am till 11 am.
```

```python
ax = plt.figure(figsize=(6, 6)).gca() # define axis
casino_data.plot.scatter(x = 'DayOfWeek', y = 'HeadCount', ax = ax, alpha = 0.1)
ax.set_title('DayOfTheWeek vs HeadCount') # Give the plot a main title
ax.set_ylabel('HeadCount')# Set text for y axis
ax.set_xlabel('DayOfWeek')
```

```python
#saturday and sunday are the busiest days at the casino
```

```python
casino_data.GameCode.unique()
```

```python
#create dictionary to convert alpha data to numeric
game = {'BA': 1, 'C4': 2, 'CR': 3, 'DH': 4, 'FP': 5, 'LI': 6, 'MI': 7, 'MQ': 8, 'MR': 9, 'MS': 10, 'ND': 11,
       'PA': 12, 'PG': 13, 'RO': 14, 'RR': 15, 'S6': 16, 'SH': 17, 'TH': 18, 'TL': 19, 'TP': 20, 'UT': 21, 'WR': 22} 
```

```python
#replacing data from dictionary 
casino_data.GameCode = [game[item] for item in casino_data.GameCode] 
```

```python
casino_data.GameCode.unique()
```

```python
import matplotlib.pyplot as plt
ax = plt.figure(figsize=(6, 6)).gca() # define axis
casino_data.plot.scatter(x = 'GameCode', y = 'HeadCount', ax = ax, alpha = 0.1)
ax.set_title('HeadCound vs GameCode') # Give the plot a main title
ax.set_ylabel('HeadCount')# Set text for y axis
ax.set_xlabel('GameCode')
```

```python
#S6: Big Six is the most played game by headcound followed by CR: Craps and TP: Texas Hold'em Poker
import seaborn as sns
```
